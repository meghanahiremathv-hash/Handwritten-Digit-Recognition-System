```python
# STEP 3 — Import Libraries
```

```python
# Import required libraries
from transformers import BertTokenizer, BertModel
import torch
```

```python
# STEP 4 — Load Pre-trained BERT Model and Tokenizer
```

```python
# Load tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Load pre-trained BERT model
model = BertModel.from_pretrained('bert-base-uncased')

print("BERT model and tokenizer loaded successfully!")
```

```python
# STEP 5 — Input the Customer Review
```

```python
# Input sentence
text = "The product is very good and easy to use."

print("Original Text:")
print(text)
```

```python
# STEP 6 — Tokenize the Sentence
```

```python
tokens = tokenizer.tokenize(text)

print("Tokens:")
print(tokens)
```

```python
# STEP 7 — Convert Tokens into Token IDs
```

```python
print("\nInterpretation:")
print("The GPT-2 model generates meaningful text based on the given starting sentence.")
```

```python
# STEP 8 — Encode the Input for BERT
```

```python
# Encode text into tensors
inputs = tokenizer(text, return_tensors="pt")

print("Encoded Input:")
print(inputs)
```

```python
# STEP 9 — Run Forward Pass Through the Model
```

```python
# Pass input through BERT model
outputs = model(**inputs)

print("Model Output:")
print(outputs)
```

```python
# STEP 10 — Display Last Hidden State Shape
```

```python
# Display hidden state shape
print("Last Hidden State Shape:")
print(outputs.last_hidden_state.shape)
```
